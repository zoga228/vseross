```python
import torch
from torchmetrics import ConfusionMatrix
```

```python
# Пример данных
y_true = torch.tensor([0, 1, 0, 1, 1, 0])  # истинные метки
y_pred = torch.tensor([0, 1, 0, 0, 1, 1])  # предсказанные метки

# Создаем объект ConfusionMatrix
confmat = ConfusionMatrix(task="binary", num_classes=2)
```

```python
# Вычисляем матрицу ошибок
matrix = confmat(y_pred, y_true)
print(matrix)
```
