Геометрический смысл

    Первая производная f′(x)f′(x) — показывает наклон касательной к графику.
    Вторая производная f′′(x)f′′(x) — показывает, насколько быстро изменяется этот наклон.

Если f′′(x)>0f′′(x)>0 на каком-то интервале, значит, функция выпукла вверх (график загибается вверх).
Если f′′(x)<0f′′(x)<0, значит, функция выпукла вниз (график загибается вниз).
