üìå –ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç?

üîπ t-SNE —Ö–æ—Ä–æ—à–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö –≤ 2D –∏–ª–∏ 3D.
üîπ –û–Ω —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç –±–ª–∏–∑–æ—Å—Ç—å —Ç–æ—á–µ–∫ –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ, —Ç–æ –µ—Å—Ç—å –ø–æ—Ö–æ–∂–∏–µ –æ–±—ä–µ–∫—Ç—ã –æ—Å—Ç–∞—é—Ç—Å—è —Ä—è–¥–æ–º.
üîπ –†–∞–±–æ—Ç–∞–µ—Ç –º–µ–¥–ª–µ–Ω–Ω–µ–µ, —á–µ–º PCA, –Ω–æ –º–æ–∂–µ—Ç –≤—ã—è–≤–ª—è—Ç—å —Å–∫—Ä—ã—Ç—ã–µ –∫–ª–∞—Å—Ç–µ—Ä—ã –≤ –¥–∞–Ω–Ω—ã—Ö.
‚ú® –ü—Ä–∏–º–µ—Ä

–î–æ–ø—É—Å—Ç–∏–º, —É –Ω–∞—Å –µ—Å—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ä—É–∫–æ–ø–∏—Å–Ω—ã—Ö —Ü–∏—Ñ—Ä (0‚Äì9). PCA –º–æ–∂–µ—Ç –Ω–µ –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–æ –æ—Ç–¥–µ–ª–∏—Ç—å —Ü–∏—Ñ—Ä—ã, –∞ t-SNE –ø–æ–∫–∞–∂–µ—Ç —á—ë—Ç–∫–∏–µ –∫–ª–∞—Å—Ç–µ—Ä—ã (–≤—Å–µ 0 —Ä—è–¥–æ–º, –≤—Å–µ 1 —Ä—è–¥–æ–º –∏ —Ç. –¥.).

```python
from sklearn.manifold import TSNE
import numpy as np

# –ü—Ä–∏–º–µ—Ä —Å–ª—É—á–∞–π–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (100 —Ç–æ—á–µ–∫, 5 –∏–∑–º–µ—Ä–µ–Ω–∏–π)
X = np.random.rand(100, 5)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ 2D
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_2d = tsne.fit_transform(X)

print(X_2d[:5])  # –í—ã–≤–µ–¥–µ–º –ø–µ—Ä–≤—ã–µ 5 —Ç–æ—á–µ–∫
```

–ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç?

üîπ MDS —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏, –∞ –Ω–µ –∏—Ö –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ.
üîπ –ï—Å–ª–∏ —É –Ω–∞—Å –µ—Å—Ç—å –º–∞—Ç—Ä–∏—Ü–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É –≥–æ—Ä–æ–¥–∞–º–∏, MDS —Å–º–æ–∂–µ—Ç –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∫–∞—Ä—Ç—É, –¥–∞–∂–µ –µ—Å–ª–∏ —É –Ω–∞—Å –Ω–µ—Ç –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç!

```python
from sklearn.manifold import MDS

mds = MDS(n_components=2, random_state=42)
X_2d = mds.fit_transform(X)

print(X_2d[:5])  # –ü–µ—Ä–≤—ã–µ 5 —Ç–æ—á–µ–∫
```

–ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç?

üîπ Isomap —Å—Ç—Ä–æ–∏—Ç –≥—Ä–∞—Ñ –±–ª–∏–∂–∞–π—à–∏—Ö —Å–æ—Å–µ–¥–µ–π –∏ —É–º–µ–Ω—å—à–∞–µ—Ç —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å, —Å–æ—Ö—Ä–∞–Ω—è—è –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫—É—é —Ñ–æ—Ä–º—É –¥–∞–Ω–Ω—ã—Ö.
üîπ –û—Ç–ª–∏—á–Ω–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –¥–ª—è –∏–∑–æ–≥–Ω—É—Ç—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Ç–æ—á–∫–∏ –Ω–∞ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ —Å—Ñ–µ—Ä—ã).

```python
from sklearn.manifold import Isomap

iso = Isomap(n_components=2, n_neighbors=5)
X_2d = iso.fit_transform(X)

print(X_2d[:5])
```

üìå –ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç?

üîπ –°–æ–≤—Å–µ–º –Ω–æ–≤—ã–π –º–µ—Ç–æ–¥, –ø–æ—Ö–æ–∂–∏–π –Ω–∞ t-SNE, –Ω–æ –Ω–∞–º–Ω–æ–≥–æ –±—ã—Å—Ç—Ä–µ–µ.
üîπ –°–æ—Ö—Ä–∞–Ω—è–µ—Ç –Ω–µ —Ç–æ–ª—å–∫–æ –ª–æ–∫–∞–ª—å–Ω—ã–µ, –Ω–æ –∏ –≥–ª–æ–±–∞–ª—å–Ω—ã–µ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞–Ω–Ω—ã—Ö.
üîπ –ü–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –±–æ–ª—å—à–∏—Ö –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö.

```python
import umap.umap_ as umap

umap_model = umap.UMAP(n_components=2)
X_2d = umap_model.fit_transform(X)

print(X_2d[:5])
```
