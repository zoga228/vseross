```python
from collections import defaultdict
```

```python
n = int(input())
X = []

for x in range(n):
    X.append(list(map(int, input().split())))

k = int(input())
C = {}

for c in range(k):
    C[c] = list(map(int, input().split()))
```

```python
points = defaultdict(list)
```

```python
def dist(x, c):
    return (sum([(x[i]-c[i])**2 for i in range(len(x))]))**0.5
```

```python
def assign(X, C):
    for x in X:
        distances = {}
        for i in range(len(C)):
            distances[i] = (dist(x, C[i])) # {1: 0.8, 2: 1.2}
        temp = min(distances.values()) # 0.8
        res = [key for key in distances if distances[key] == temp] # {1: 0.8, 2: 1.2}
        points[res[0]].append(x)

    return points
```

```python
def update(C, points):
    new_c = {}
    for c in C:
        new_c[c] = [sum(x[0] for x in points[c]) / len(points[c]),
                    sum(x[1] for x in points[c]) / len(points[c])]
    points = defaultdict(list)
    return new_c
```

```python
def fit(X, C):
    for _ in range(1000):
        points = assign(X, C)
        C = update(C, points)
    return C
```

```python
print(fit(X, C))
```
