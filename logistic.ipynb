```python
import matplotlib.pyplot as plt
from math import log
```

```python
class LogisticReg:
    def __init__(self, lr=0.01, epochs=100):
        self.lr = lr
        self.epochs = epochs
        self.w1 = 0
        self.w2 = 0
        self.b = 0

    def predict(self, X):
        self.preds = []
        for x1, x2 in X:
            f = self.w1*x1 + self.w2*x2 + self.b
            self.preds.append(1/(1+2.71828**(-f)))
        return self.preds

    def fit(self, X, y):
        n = len(y)
        for _ in range(self.epochs):
            predictions = self.predict(X)
            dw1 = sum(((predictions[i] - y[i])*X[i][0]) for i in range(n))/n
            dw2 = sum(((predictions[i] - y[i])*X[i][1]) for i in range(n))/n
            db = sum((predictions[i] - y[i]) for i in range(n))/n

            self.w1 -= self.lr*dw1
            self.w2 -= self.lr*dw2
            self.b -= self.lr*db
```

```python
X = [[1.7606, -1.6402],
     [1.1556,  0.4826],
     [0.5243, -0.6675],
     [1.0260, -0.8620],
     [0.5337, -0.6402],
     [0.6692,  1.8840],
     [1.0895,  0.7105],
     [0.3719,  1.6948],
     [2.3394, -2.2632],
     [0.7709,  0.9314],
     [1.0347,  1.1996],
     [2.2541, -2.3520],
     [0.8301,  0.4806]]
y = [0, 1, 0, 0, 0, 1, 1, 1, 0, 1, 1, 0, 1]

model = LogisticReg(lr=0.01, epochs=1000)
model.fit(X, y)

print("w1:", model.w1)
print("w2:", model.w2)
print("b:", model.b)
```

```python
w1 = model.w1
w2 = model.w2
b = model.b
x1 = []
for i in range(0, 4, 1):
    x1.append(i)
x2 = [(-(w1 * x1[i] + b) / w2) for i in range(len(x1))]

print(x1, x2)
```

```python
xx = []
xy = []
for q, z in X:
    xx.append(q)
    xy.append(z)

y = [0, 1, 0, 0, 0, 1, 1, 1, 0, 1, 1, 0, 1]

plt.scatter(xx, xy, c=y, cmap='viridis')
plt.plot(x1, x2, color='red', label='Граница решения')
plt.xlabel("Признак 1")
plt.ylabel("Признак 2")
plt.legend()
plt.show()
```
