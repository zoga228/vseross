```python
import torch
from torchmetrics import ROC
import matplotlib.pyplot as plt
```

```python
# Пример данных
y_true = torch.tensor([0, 1, 0, 1, 1, 0])  # истинные метки
y_scores = torch.tensor([0.1, 0.9, 0.4, 0.6, 0.8, 0.3])  # вероятности класса 1
```

```python
# Создаем объект ROC
roc = ROC(task="binary")

# Вычисляем FPR, TPR и пороги
fpr, tpr, thresholds = roc(y_scores, y_true)
```

```python
# Строим график ROC Curve
plt.plot(fpr, tpr, label="ROC Curve")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
from torchmetrics import AUROC

# Создаем объект AUROC
auroc = AUROC(task="binary")

# Вычисляем AUC-ROC
auc_score = auroc(y_scores, y_true)
print(f"AUC-ROC: {auc_score}")
```
