```python
import numpy as np

# 1. Зададим матрицу A
A = np.array([[1, 2],
              [3, 4]])
print("Матрица A:")
print(A)

# 2. Вычисляем SVD разложение
# Функция возвращает три объекта:
# U – матрица левых сингулярных векторов,
# S – массив сингулярных значений (не диагональная матрица),
# VT – транспонированная матрица правых сингулярных векторов.
U, S, VT = np.linalg.svd(A)

print("\nМатрица U:")
print(U)

print("\nСингулярные значения (массив):")
print(S)

print("\nМатрица VT:")
print(VT)

# 3. Если нужно получить диагональную матрицу Σ, можно сделать так:
Sigma = np.zeros((A.shape[0], A.shape[1]))
np.fill_diagonal(Sigma, S)

print("\nДиагональная матрица Σ:")
print(Sigma)
```
