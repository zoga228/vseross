```python
import torch
import torch.nn as nn
import torch.nn.init as init

linear_layer = nn.Linear(20, 10)

# Инициализация весов с помощью Xavier Uniform
init.xavier_uniform_(linear_layer.weight)
if linear_layer.bias is not None:
    init.zeros_(linear_layer.bias)

print("Weights after Xavier initialization:")
print(linear_layer.weight)
```

```python
import torch
import torch.nn as nn
import torch.nn.init as init

conv_layer = nn.Conv2d(in_channels=3, out_channels=16, kernel_size=3, padding=1)

# Инициализация весов с помощью He Normal (Kaiming Normal)
init.kaiming_normal_(conv_layer.weight, mode='fan_in', nonlinearity='relu')
if conv_layer.bias is not None:
    init.zeros_(conv_layer.bias)

print("Weights after He initialization:")
print(conv_layer.weight)
```

```python
def weights_init(m):
    if isinstance(m, nn.Linear): # проверка на принадлежность классу
        init.xavier_uniform_(m.weight)
        if m.bias is not None:
            init.zeros_(m.bias)
    elif isinstance(m, nn.Conv2d):
        init.kaiming_normal_(m.weight, mode='fan_in', nonlinearity='relu')
        if m.bias is not None:
            init.zeros_(m.bias)

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, padding=1)
        self.relu = nn.ReLU()
        self.fc = nn.Linear(16 * 32 * 32, 10)  # допустим, входное изображение 32x32

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x

model = SimpleCNN()
model.apply(weights_init)
```
